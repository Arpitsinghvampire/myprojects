```python
#here we import the required dependencies
#we import tensorflow , the keras and the dataset

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import fashion_mnist
```

```python
#now we split the data into train and the test set
(x_train,y_train),(x_test,y_test)=fashion_mnist.load_data()
#this creates the x_train and the y_train
#when we split the data using this method what happens is that the training set will be shuffled before the split
```

```python
#now here we can check the dimensionality of the train and the test set

#other thing to note is that the x_train ,y_train ,x_test ,y_test from this method would be in the form of numpy array

#here we print the sizes
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#now after differentiating between the train and the test data , we now create the cnn model
#here we will use the Sequential api in keras to create the model
model=keras.Sequential(
    [
        keras.Input((28,28,1)),
        layers.Conv2D(32,3,padding='valid',activation='relu'),
        layers.MaxPooling2D(pool_size=(2,2)),
        layers.Conv2D(64,3,activation='relu'),
        layers.MaxPooling2D(),
        layers.Conv2D(128,3,activation='relu'),
        layers.Flatten(),
        layers.Dense(64,activation='relu'),
        layers.Dense(10),

    ]
)
```

```python
#once we have created our model , we need to get the summary of the model
model.summary()
```

```python
#here we want to compile the model after creating the model
model.compile(
    optimizer='adam',  # Choose your optimizer, e.g., 'adam'
    loss=keras.losses.MeanSquaredError(),  # Specify the loss function
    metrics=['accuracy']  # Specify the evaluation metric(s), e.g., 'accuracy'
)
```

```python
#after compiling the model we want to fit the data
model.fit(x_train,y_train,batch_size=96,epochs=3,verbose=1)
#the parameters are x_train ,y_train , the batch size , the number of eepochs ,and then how we represent the how we represent the progress

```

```python
#now after fitting the train set to the model we would need to evaluate the model
model.evaluate(x_test,y_test,batch_size=32,verbose=1)
#this takes in the parameter as x_test , y_test , the batch size and the verbose
#the accuracy is less
```

```python
import numpy as np
# now we want to see the predictions
predictions=model.predict(x_test) #this will give us predictions for the x_test
print(predictions[0]) #this will give us the logits , we would need to convert it to softmax function
#to get the required class
print(np.argmax(predictions[0])) #this will give us the predicted class of the test case
```
